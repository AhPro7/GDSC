# Splite the Data

```python
# split train data
X_train_df = train_df.drop('Activity',axis=1)
Y_train_df = train_df['Activity']

# Split test for evalution
X_test_df = test_df.drop('Activity',axis=1)
Y_test_df = test_df['Activity']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X_train_df,Y_train_df,test_size=0.2,random_state=145)
```

# Simple ML 

```python
# Machine learning Model
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
LDA = LinearDiscriminantAnalysis()
LDA.fit(x_train,y_train)
```

```python
y_perd = LDA.predict(X_test_df)
from sklearn.metrics import classification_report,accuracy_score
print(classification_report(y_perd,Y_test_df))
```

# DL model

```python
# fixed
classes_num = len(Y_train_df.unique()) # 6 # train or test
input_size = len(X_train_df.columns) # 562 = number of features # must be the train 

Dl_model = tf.keras.Sequential([

    tf.keras.layers.Dense(input_size,input_shape=(input_size,)),
    tf.keras.layers.Dense(400,activation='relu'),
    tf.keras.layers.Dense(300,activation='relu'),
    tf.keras.layers.Dense(200,activation='relu'),
    tf.keras.layers.Dense(100,activation='relu'),
    tf.keras.layers.Dense(50,activation='relu'),
    tf.keras.layers.Dense(25,activation='relu'),
    tf.keras.layers.Dense(10,activation='relu'),
    tf.keras.layers.Dense(classes_num,activation='softmax')

])
```

```python
Dl_model.summary()
```

```python
# compile
Dl_model.compile(
    optimizer = tf.keras.optimizers.Adam(),
    loss = tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics = ['accuracy']
)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
y_train_enc =  le.fit_transform(y_train)
y_test_enc = le.transform(y_test)
Y_test_df_enc = le.transform(Y_test_df)
```

```python
his = Dl_model.fit(
    x = x_train,y = y_train_enc,

    validation_data =(x_test,y_test_enc),
    
    epochs=20,
    
    
)
```

```python
Dl_model.evaluate(X_test_df,Y_test_df_enc)
```

# Get prdicatoins from Dl model

```python
pred = Dl_model.predict(X_test_df)[0]
print('Trust ratio of each class ==> ',pred)
print('Most trusted class index ==> ',pred.argmax())
print('Most trusted class is ==> ',le.inverse_transform([pred.argmax()]))
print('Most trusted class ratio is ==> ',pred.max()*100,'%')
print('Real class is ==> ',Y_test_df[0])
```

```python
y_perd_dl = Dl_model.predict(X_test_df)
y_perd_dl_class = []
for i in y_perd_dl:
    y_perd_dl_class.append(i.argmax())


from sklearn.metrics import classification_report,accuracy_score

print(classification_report(y_perd_dl_class,Y_test_df_enc))
```

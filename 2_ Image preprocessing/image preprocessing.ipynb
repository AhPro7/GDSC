```python
import matplotlib.pyplot as plt

import numpy as np

import tensorflow as tf
import cv2

from tensorflow.keras import layers
```

```python
from keras.datasets import fashion_mnist
(training_data, _), (_, _) = fashion_mnist.load_data()
```

```python
def visualize_input(img, ax):
    ax.imshow(img, cmap='gray')
    width, height = img.shape
    thresh = img.max()/2.5
    for x in range(width):
        for y in range(height):
            ax.annotate(str(round(img[x][y],2)), xy=(y,x),
            horizontalalignment='center',
            verticalalignment='center',
            color='white' if img[x][y]<thresh else 'black')
```

```python
plt.imshow(training_data[3343],cmap='gray')
```

```python
fig = plt.figure(figsize = (12,12))
ax = fig.add_subplot(111)
visualize_input(training_data[3343], ax)
```

```python
import matplotlib.pyplot as plt

# to compare between images

def plot_comparison (original, filtered, title_filtered):
   fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 8))
   ax1.imshow(original)
   ax1.set_title('original')
   # ax1. axis('off')
   ax2.imshow(filtered)
   ax2.set_title(title_filtered)
   # ax2.axis('off')
```

```python
# main image for playing
# we have many ways but i will use simplest 

main_image = cv2.imread('Images/logo.png')
main_image = cv2.cvtColor(main_image, cv2.COLOR_BGR2RGB)

# how computer see image
print(type(logo2))
print(logo2.dtype)
print(logo2.shape)

plt.imshow(main_image)
```

# Resize image

```python
resized = cv2.resize(main_image, (200, 200))
plot_comparison(main_image,resized,'resized')
```

# Gaussian Blur


```python
Gaussian = cv2.GaussianBlur(main_image, (31, 31), 0)
plot_comparison(main_image,Gaussian,'Gaussian Blur')
```

# Converting color images to grayscale to reduce computation complexity

```python
gray = cv2.cvtColor(main_image, cv2.COLOR_RGB2GRAY)
plot_comparison(main_image,gray,'COLOR_RGB2GRAY')
print(gray.shape)
```

# Edge Detection

```python
edges = cv2.Canny(main_image,100,110)

plot_comparison(main_image,edges,'edges cany')
```

```python
sobel = cv2.Sobel(main_image, -1, 1, 1)

plot_comparison(main_image,edges,'sobel')
```
